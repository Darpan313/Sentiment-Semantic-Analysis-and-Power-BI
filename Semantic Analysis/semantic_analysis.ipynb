```python
import pandas as pd
import math
```

```python
df_news = pd.read_csv('clean_news.csv') #cleaned files in 2nd Assignment...
```

```python
numOfDocs = df_news.shape[0]
print("Number of documents: ",df_news.shape[0])
```

```python
df_news = df_news.loc[:, ~df_news.columns.str.contains('^Unnamed')] 
```

```python
df_news['news'] = df_news['title'] +' '+ df_news['description'] +' '+ df_news['content']
```

```python
df_news = df_news[['news']]
```

```python
df_news
```

```python
searchKeys = ["Canada","University","Dalhousie University","Halifax","Canada Education"]
df_details = pd.DataFrame(columns = ['Search Query','Document containing term(df)', 'Total Documents(N)/ number of documents term appeared (df)','Log10(N/df)'])
df_details['Search Query'] = searchKeys
```

```python
listDfs = []
listNbyDfs = []
listLogNbyDfs = []
for w in searchKeys:
    count = 0
    w = w.lower()
    for index,row in df_news.iterrows():
        news = row['news']
        if w in news:
            count+=1
            
    listDfs.append(count)
    listNbyDfs.append(str(numOfDocs)+"/"+str(count))
    if(count!=0):
        listLogNbyDfs.append(math.log10(numOfDocs/count))
    else:
        listLogNbyDfs.append("NA")

df_details['Document containing term(df)'] = listDfs
df_details['Total Documents(N)/ number of documents term appeared (df)'] = listNbyDfs
df_details['Log10(N/df)'] = listLogNbyDfs
```

```python
print("Number of documents: ",df_news.shape[0])
df_details
```

```python
search = 'Canada'
df_detail_news = pd.DataFrame(columns = ["Term "+search+" in 160 docs","Total words(m)","Frequency(f)"])
print("Search term: ",search)
search_lower = search.lower()
listTotalWordCount = []
listSearchWordCount = []
listArticle = []
for index,row in df_news.iterrows():
    count=0
    news = row['news']
    if search_lower in news:
        tokenized = news.split(" ")
        total_words = len(tokenized)
        for word in tokenized:
            if word == search_lower:
                count+=1
        listTotalWordCount.append(total_words)
        listSearchWordCount.append(count)
        listArticle.append("Article #"+ str(index)) 
df_detail_news["Term "+search+" in 160 docs"]=listArticle
df_detail_news["Total words(m)"]=listTotalWordCount
df_detail_news["Frequency(f)"]=listSearchWordCount
df_detail_news
```

```python
df_detail_news['f/m'] = df_detail_news.apply(lambda x: x['Frequency(f)']/x['Total words(m)'],axis=1)
```
