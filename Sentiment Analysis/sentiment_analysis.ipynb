```python
import pandas as pd
import collections as coll
import csv
```

```python
ignore = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
```

```python
listOfPosWords = []
with open('positive-words.txt') as f:  #positive-words.txt source is refrenced in text file...
    for line in f:
        if not line.startswith(';'):
            line = line.strip('\n')
            listOfPosWords.append(line)
listOfPosWords = listOfPosWords[1:] 
```

```python
listOfNegWords = []
with open('negative-words.txt') as f:  #negative-words.txt source is refrenced in text file...
    for line in f:
        if not line.startswith(';'):
            line = line.strip('\n')
            listOfNegWords.append(line)
listOfNegWords = listOfNegWords[1:] 
```

```python
df_tweets = pd.read_csv("clean_tweets.csv")
df_tweets = df_tweets.loc[:, ~df_tweets.columns.str.contains('^Unnamed')] #https://stackoverflow.com/questions/43983622/remove-unnamed-columns-in-pandas-dataframe/43983654
df_tweets = df_tweets[['Tweet']]
```

```python
df_tweets
```

```python
globalPosWordsInTweets = []
globalNegWordsInTweets = []
sentimentlist = []
listOfListPos = []
listOfListNeg = []
for index,row in df_tweets.iterrows():
    tweet = row['Tweet']
    posWordInTweet = []
    negWordInTweet = []
    cntPositive = 0
    cntNegative = 0
    sentiment = "neutral"
    listOfWords = []
    tokenized = tweet.split(" ")
    for word in tokenized:
        if word not in ignore:
            listOfWords.append(word)
    BOW = coll.Counter(listOfWords)
    #print(BOW)
    for word in BOW:
        #print(word)
        if word in listOfPosWords:
            posWordInTweet.append(word)
            globalPosWordsInTweets.append(word)
            cntPositive+=BOW[word]
        if word in listOfNegWords:
            negWordInTweet.append(word)
            globalNegWordsInTweets.append(word)
            cntNegative+=BOW[word]    
        if(len(posWordInTweet)>len(negWordInTweet)):
             sentiment = "positive"
        elif(len(posWordInTweet)<len(negWordInTweet)):
            sentiment =  "negative"
        elif((len(posWordInTweet) == len(negWordInTweet))):
            sentiment =  "neutral"
    sentimentlist.append(sentiment)
    listOfListPos.append(posWordInTweet)
    listOfListNeg.append(negWordInTweet)
```

```python
df_tweets['positive_match']=listOfListPos
df_tweets['negative_match']=listOfListNeg
df_tweets['sentiment']=sentimentlist
```

```python
c = coll.Counter(globalNegWordsInTweets)
with open('NegativeWords.csv','w',newline="") as csv_file:
    writer = csv.writer(csv_file)
    for x in c:
        key = x
        value = c[key]
        writer.writerow([key,value])
    
```
